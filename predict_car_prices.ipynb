```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_csv = pd.read_csv("training.csv")
train_x_a = train_csv.drop('price', 1)
train_y = train_csv["price"].to_frame()


#train_x_a ALL COLUMNS EXCEPT TARGET COLUMN " PRİCE "
#train_y  ONLY TARGET COLUMN
```

```python
train_x_a
```

```python
train_y
```

```python
indexes = train_y['price'].index[train_y['price'].apply(np.isnan)]
indexes
#getting indexes of nan values so we can drop them from target 
```

```python
train_x_a.drop(indexes, inplace=True)
train_y.drop(indexes, inplace = True)
```

```python
train_x_a
```

```python
train_y
```

```python
nan_columns = list(train_x_a.columns[train_x_a.isna().any()])
nan_columns
#getting columns that have nan values 
```

```python
train_x_a = train_x_a.fillna(train_x_a.mean())
train_x_a
#filling nan values with their mean
```

```python
cols = train_x_a.columns
num_cols = train_x_a._get_numeric_data().columns
categorical_columns = list(set(cols) - set(num_cols))
categorical_columns
#getting categorical columns and numeric columns 
# i use categorical columns for one hot encoding , numeric columns for scaling
```

```python
train_x_a
```

```python
train_x_d = pd.get_dummies(train_x_a)
#one hot encoding
```

```python
train_x_d
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(train_x_d[num_cols])
train_x_d[num_cols] =scaler.transform(train_x_d[num_cols])
train_x_d
#scaling 

```

```python
train_x_d[num_cols].hist(bins=50, figsize=(20,15))
plt.show()
```

```python
combined = pd.concat([train_y, train_x_d[num_cols]], axis=1) 
combined
#adding target column and other numeric columns into one dataframe to colculate correlation
```

```python
correlations = combined.corr()
top_5_corr = list(correlations['price'].index[1:6])
top_5_corr
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
# Train model
linearRegression = LinearRegression()
linearRegression.fit(train_x_d, train_y)

train_pred = linearRegression.predict(train_x_d) 
#Predict on training data nad calculate train MSE.
mse = mean_squared_error(train_y, train_pred)
print(mse)
```

```python
from sklearn.model_selection import cross_val_score


cv_lr_model = LinearRegression()
train_mse_score = -cross_val_score(cv_lr_model, train_x_d, train_y, cv=5, scoring='neg_mean_squared_error')
print(train_mse_score) 
# Get the negative of the result to get MSE.
```

```python
average_mse = sum(train_mse_score) / len(train_mse_score) #Get Mean MSE
print(average_mse)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
poly_dataset = poly.fit_transform(train_x_d) # PolynomialFeatures transforming returns an array so it has be formed into a DataFrame.
train_x_m = pd.DataFrame(poly_dataset)
train_x_m
```

```python
lr_model_poly = LinearRegression()
poly_mse = -cross_val_score(lr_model_poly, train_x_m, train_y, cv=5, scoring='neg_mean_squared_error')
poly_mse
```

```python
average_mse = sum(poly_mse) / len(poly_mse)
print(average_mse) #Error is much higher than before polynomial transformation, probably due to overfitting.
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso()
lasso_mse = -cross_val_score(lasso, train_x_m, train_y, cv=5, scoring='neg_mean_squared_error')
```

```python
average_mse = sum(lasso_mse) / len(lasso_mse)
print(average_mse) #Error is lower than linear regression model, regularization is improviing our model.
```

```python
all_scores = []

for alpha in range(1, 3001, 10): # Go through all alphas from 1 to 3000
    lasso = Lasso(alpha = alpha) #Create new Lasso model
    lasso_mse = -cross_val_score(lasso, train_x_m, train_y, cv=5, scoring='neg_mean_squared_error')
    average_mse = sum(lasso_mse) / len(lasso_mse)
    all_scores.append(average_mse) # Store score

score_series = pd.Series(all_scores, index=range(1, 3001, 10)) # Transform scores in to series
score_series
#min_error = score_series.min() #Get minimum score
#min_alpha = score_series.idxmin() #Get minimum index which represents the alpha
```
